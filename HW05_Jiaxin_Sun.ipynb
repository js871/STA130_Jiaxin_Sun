When calculating a p-value, we start by assuming that the null hypothesis is true because we need a reference point to compare our observed data against. The null hypothesis represents the "status quo" or the assumption that there's no effect or no difference. By imagining a world where the null hypothesis is true, we can create a sampling distribution that shows what outcomes we would expect for the test statistic (like a sample mean or proportion) if the null hypothesis were indeed correct.

Once we have this sampling distribution under the null hypothesis, we can see how likely our observed data (or more extreme values) are compared to this distribution. If the observed data fall far into the tail of this distribution, it suggests that the result is unusual if the null hypothesis is true, meaning we have evidence to reject the null.

In short, assuming the null hypothesis is true helps us understand the "expected" behavior of the data, and the p-value tells us how surprising our actual data is compared to that expectation.

When you start with a null hypothesis, you're basically saying, "Let’s assume there’s no effect, no difference, or no relationship between the variables." This is like your baseline assumption.

Once you assume the null hypothesis is true, you imagine what all the possible outcomes could be if this assumption is correct. This is where the **sampling distribution** comes in—it's a graph or model that shows how likely different results are when the null hypothesis holds true.

- If your test statistic (for example, a sample mean or proportion) falls in the **center** of this distribution, it means your result is typical and expected under the null hypothesis.
- But if your test statistic falls in the **tails** of this distribution (far from the center), it means your result is very **unusual** under the null hypothesis.

Then,the **p-value** tells you **how unusual** your result is, given the null hypothesis is true. It is the **probability** of getting a result as extreme as (or more extreme than) what you observed, assuming the null hypothesis is correct.

But, **small p-value** means that the result you observed is very unlikely to occur if the null hypothesis were true. 

- For example, imagine flipping a coin 100 times and getting 90 heads. The null hypothesis would say the coin is fair (50/50), but getting 90 heads would be an **extremely unlikely** outcome under that assumption.
- The **smaller** the p-value, the more **unlikely** it is to get your result if the null hypothesis is correct. This suggests that maybe the null hypothesis is wrong.

Additionaly,the smaller the p-value, the **stronger** the evidence **against** the null hypothesis. When the p-value is very small, it indicates that the observed data doesn’t fit well with the assumption that the null hypothesis is true. Therefore, it makes the null hypothesis seem less and less likely to be correct, which is why people say it makes the null hypothesis look "ridiculous."

To clarify:
- **Small p-value** = Unlikely result under the null hypothesis = Stronger evidence to reject the null hypothesis.

# **Tea Experiment: Statistical Analysis Report**

## **Introduction**

This report presents the statistical analysis of an experiment where 80 STA130 students were asked to identify whether milk or tea was poured first in a cup of tea. Out of the 80 students, 49 correctly identified the pouring order. The goal of the experiment is to test whether students are capable of identifying the order better than by random guessing, using statistical methods to assess the validity of this claim.

## **Population and Sample**

- **Population**: The population of interest includes all STA130 students who could potentially participate in this experiment and identify whether tea or milk was poured first.
  
- **Sample**: A random sample of 80 STA130 students was selected for the experiment. Each student tasted one cup of tea and was asked to identify the pouring order. Out of these 80 students, 49 provided correct answers.

- **Parameter of Interest**: The parameter of interest is the proportion, \( p \), of students who can correctly identify whether tea or milk was poured first.

## **Statistical Hypotheses**

### **Formal Null Hypothesis ( \( H_0 \) )**

The null hypothesis, \( H_0 \), assumes that the students are merely guessing the pouring order, meaning the probability of correctly identifying the order is 50%. This can be formally stated as:
\[
H_0: p = 0.5
\]
Where \( p \) is the true proportion of correct identifications.

### **Informal Interpretation of \( H_0 \)**

In simple terms, \( H_0 \) implies that students are guessing randomly, and thus, they have a 50% chance of guessing the order correctly.

### **Alternative Hypothesis ( \( H_A \) )**

The alternative hypothesis, \( H_A \), suggests that the proportion of correct answers is greater than 50%, implying that students are capable of identifying the order better than random guessing. Formally, this is expressed as:
\[
H_A: p > 0.5
\]
This is a one-sided hypothesis test, where we are only interested in whether the proportion is significantly greater than 0.5.

## **Quantitative Analysis**

### **Methodology**

To test the hypotheses, we perform a one-sided **z-test for proportions**. The test evaluates whether the sample proportion \( \hat{p} \), which is 0.6125 (49 out of 80 correct), is significantly greater than 0.5 under the null hypothesis.

1. **Sample Proportion**:
   \[
   \hat{p} = \frac{49}{80} = 0.6125
   \]

2. **Standard Error Calculation**: The standard error of the sample proportion is calculated using the formula:
   \[
   SE = \sqrt{\frac{p_0(1 - p_0)}{n}} = \sqrt{\frac{0.5(1 - 0.5)}{80}} = 0.0559
   \]
   
3. **Z-score Calculation**: The z-score is calculated using the following formula:
   \[
   z = \frac{\hat{p} - p_0}{SE} = \frac{0.6125 - 0.5}{0.0559} = 2.01
   \]
   
4. **P-value Calculation**: For a one-sided test, we find the p-value by calculating the area to the right of the z-score:
   \[
   p\text{-value} = 1 - \text{CDF}(2.01) = 0.0222
   \]
   
5. **Significance Level**: We use a significance level of \( \alpha = 0.05 \).

### **Interpretation of Results**

- **Z-score**: The z-score is **2.01**, which measures how many standard deviations the sample proportion is from the null hypothesis proportion of 0.5.
  
- **P-value**: The p-value is **0.0222**, which represents the probability of observing a sample proportion as extreme as 0.6125 or more extreme, assuming the null hypothesis is true.

Since the p-value (0.0222) is **less than** the significance level \( \alpha = 0.05 \), we **reject the null hypothesis**. This suggests that the students are likely better than random guessing at identifying the order of pouring.

## **Conclusion**

Based on the z-test for proportions, the analysis provides evidence to reject the null hypothesis at the 5% significance level. The p-value of 0.0222 indicates that it is unlikely that the observed result (49 correct out of 80) occurred due to random guessing alone. Therefore, we conclude that there is statistical evidence to support the claim that the students can correctly identify the pouring order better than by chance.

```python
import numpy as np
import scipy.stats as stats

# Sample Data
n = 80  # Total number of students
x = 49  # Number of students who correctly identified the order
p_hat = x / n  # Sample proportion
p_0 = 0.5  # Null hypothesis proportion (random guessing)

# Hypothesis test parameters
alpha = 0.05  # Significance level

# Step 1: Calculate the z-score
# z = (p_hat - p_0) / sqrt((p_0 * (1 - p_0)) / n)
std_error = np.sqrt(p_0 * (1 - p_0) / n)
z_score = (p_hat - p_0) / std_error

# Step 2: Calculate the p-value
# Since it's a one-tailed test, we use 1 - CDF of z_score
p_value = 1 - stats.norm.cdf(z_score)

# Step 3: Conclusion
print(f"Sample proportion (p̂): {p_hat:.4f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print(f"Reject the null hypothesis (p < {alpha}).")
    print("There is evidence that the students can distinguish the order better than by chance.")
else:
    print(f"Fail to reject the null hypothesis (p >= {alpha}).")
    print("There is no strong evidence that the students can distinguish the order better than by chance.")
```
