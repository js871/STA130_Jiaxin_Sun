```python
# HW0204

from collections import defaultdict
import random

# Sample text data
text = "hello world hello chatbot world hello chatbot"
words = text.split()

# Initialize dictionaries
word_used = defaultdict(int)
next_word = defaultdict(lambda: defaultdict(int))

# Build the Markov model
for i, word in enumerate(words[:-1]):
    word_used[word] += 1
    next_word[word][words[i+1]] += 1

# Function to generate text using the Markov model
def generate_text(start_word, length=10):
    if start_word not in next_word:
        return "Start word not in model."

    current_word = start_word
    output = [current_word]

    for _ in range(length - 1):
        next_words = next_word[current_word]
        if not next_words:
            break
        
        # Choose the next word based on probabilities
        next_word_total = sum(next_words.values())
        rand_choice = random.randint(1, next_word_total)
        
        cumulative = 0
        for word, count in next_words.items():
            cumulative += count
            if rand_choice <= cumulative:
                current_word = word
                output.append(current_word)
                break
    
    return ' '.join(output)

# Example usage
start_word = 'hello'
print(generate_text(start_word, length=10))


# Explanation ：
# Building the Model:
# word_used: Tracks how often each word appears.
# next_word: Tracks how often each word is followed by each possible next word.
# Generating Text:
# Start Word: Begin with start_word.
# Next Word: Choose the next word based on the probabilities derived from next_word.
# Repeat: Continue until the desired length is reached or no more next words are available.
```

```python
#HW0205

# 2. Extension 1: Character-Specific Markov Chains
# Code Extension:


# Sample text data with characters
text = "hello world hello chatbot world hello chatbot"
characters = text.split()  # Treat each word as a character for this example

# Initialize dictionaries for characters
char_used = defaultdict(int)
next_char = defaultdict(lambda: defaultdict(int))

# Build the character-specific Markov model
for i, char in enumerate(characters[:-1]):
    char_used[char] += 1
    next_char[char][characters[i+1]] += 1

# Function to generate text with character-specific Markov model
def generate_char_text(start_char, length=10):
    if start_char not in next_char:
        return "Start character not in model."

    current_char = start_char
    output = [current_char]

    for _ in range(length - 1):
        next_chars = next_char[current_char]
        if not next_chars:
            break
        
        # Choose the next character based on probabilities
        next_char_total = sum(next_chars.values())
        rand_choice = random.randint(1, next_char_total)
        
        cumulative = 0
        for char, count in next_chars.items():
            cumulative += count
            if rand_choice <= cumulative:
                current_char = char
                output.append(current_char)
                break
    
    return ' '.join(output)

# Explanation:
# Data Preparation:

# The text string is split into words, treating each word as a "character" for simplicity in this example.
# Initialization:

# char_used: Keeps track of the frequency of each word (character).
# next_char: Maps each word (character) to the frequency of the words (characters) that follow it.
# Building the Model:

# Loop through the list of words, updating char_used and next_char to count how often each word is followed by another word.
# Text Generation:

# generate_char_text function starts with a given word and generates a sequence of words by selecting the next word based on probabilities derived from next_char.



# 3. Extension 2: Using Bigrams for Dependency
# Code Extension:

# Sample text data
text = "hello world hello chatbot world hello chatbot"
words = text.split()

# Initialize dictionaries for bigrams
bigram_counts = defaultdict(lambda: defaultdict(int))

# Build the bigram Markov model
for i in range(len(words) - 1):
    bigram_counts[words[i]][words[i+1]] += 1

# Function to generate text using the bigram Markov model
def generate_bigram_text(start_word, length=10):
    if start_word not in bigram_counts:
        return "Start word not in model."

    current_word = start_word
    output = [current_word]

    for _ in range(length - 1):
        next_words = bigram_counts[current_word]
        if not next_words:
            break
        
        # Choose the next word based on probabilities
        next_word_total = sum(next_words.values())
        rand_choice = random.randint(1, next_word_total)
        
        cumulative = 0
        for word, count in next_words.items():
            cumulative += count
            if rand_choice <= cumulative:
                current_word = word
                output.append(current_word)
                break
    
    return ' '.join(output)


# The text string is split into words, just as before.
# Initialization:
# bigram_counts: Tracks how often each word is followed by another specific word (a bigram).

# Building the Model:
# Loop through the list of words, counting how often each word is followed by another word to build the bigram model.

# Text Generation:
# generate_bigram_text function starts with a given word and generates a sequence of words based on bigram probabilities.

# Key Differences:
# Character-Specific Markov Chains:

# Treats each word as a "character" and models the probability of one word following another. It doesn't use word pairs but rather single words.
# Bigram Markov Chains:

# Uses pairs of consecutive words (bigrams) to model the probabilities. It captures more context by considering the previous word in the sequence when predicting the next word.
# How to Use:
# Character-Specific Markov Chains:

# Useful if you want to model text based on single-word transitions without considering the word context.
# Bigram Markov Chains:

# Provides a richer context by considering pairs of words, which often results in more coherent text generation.


# HERE is my chatGPT chat recording:
# https://chatgpt.com/share/66ea2353-7bbc-8005-a31b-cf8e7201643a
```
